```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
# Load data
df = pd.read_csv('/content/Groceries_dataset.csv')  # Replace with your file path
df.head()
```

```python
# Convert 'Date' to datetime
df['Date'] = pd.to_datetime(df['Date'])

# Check for missing values
print(df.isnull().sum())

# Check number of unique customers and items
print("Unique Customers:", df['Member_number'].nunique())
print("Unique Items:", df['itemDescription'].nunique())
```

```python
# Group items bought per transaction (customer + date)
basket = df.groupby(['Member_number', 'Date'])['itemDescription'].apply(list).reset_index()
basket.head()
```

```python
from mlxtend.preprocessing import TransactionEncoder

te = TransactionEncoder()
te_ary = te.fit(basket['itemDescription']).transform(basket['itemDescription'])
df_encoded = pd.DataFrame(te_ary, columns=te.columns_)
df_encoded.head()
```

```python
frequent_itemsets = apriori(df_encoded, min_support=0.02, use_colnames=True)
frequent_itemsets.sort_values(by='support', ascending=False).head()
```

```python
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
rules.sort_values(by='confidence', ascending=False).head(10)
```

```python
# Show the strongest rules
print("Top 10 Association Rules:")
rules[['antecedents', 'consequents', 'support', 'confidence', 'lift']].head(10)
```

```python
# Displaying insights
print("\n📌 Marketing Insights:\n")

for i, row in rules.head(5).iterrows():
    antecedents = ', '.join(list(row['antecedents']))
    consequents = ', '.join(list(row['consequents']))
    print(f"🛍️ If a customer buys [{antecedents}], they are likely to also buy [{consequents}]")
    print(f"   → Confidence: {row['confidence']:.2f}, Lift: {row['lift']:.2f}\n")

print("📈 Recommendations:")
print("- Bundle frequently bought items together in promotions.")
print("- Recommend associated products in online and in-store shopping.")
print("- Highlight high-lift rules in email campaigns or loyalty programs.")
```

```python
plt.figure(figsize=(8,6))
plt.scatter(rules['support'], rules['confidence'], alpha=0.6, c=rules['lift'], cmap='viridis')
plt.colorbar(label='Lift')
plt.title('Association Rules: Support vs Confidence')
plt.xlabel('Support')
plt.ylabel('Confidence')
plt.grid(True)
plt.show()
```

```python
top_items = df['itemDescription'].value_counts().head(10)
plt.figure(figsize=(10,6))
sns.barplot(x=top_items.values, y=top_items.index, palette='viridis')
plt.title('Top 10 Most Purchased Items')
plt.xlabel('Purchase Count')
plt.ylabel('Item')
plt.show()
```

```python
from collections import Counter
import matplotlib.pyplot as plt

# Flatten all consequents into one list
consequent_items = [item for sublist in rules['consequents'] for item in sublist]

# Count frequency of each item
item_counts = Counter(consequent_items)

# Get top 5
top_items = item_counts.most_common(5)
labels, sizes = zip(*top_items)

# Plot donut chart
fig, ax = plt.subplots(figsize=(8, 8))
wedges, texts, autotexts = ax.pie(
    sizes,
    labels=labels,
    autopct='%1.1f%%',
    startangle=140,
    wedgeprops=dict(width=0.4),
    textprops=dict(color="black")
)

plt.setp(autotexts, size=10, weight="bold")
ax.set_title("Top 5 Consequent Items in Association Rules", fontsize=14)
plt.show()
```

```python
# 📦 Install required libraries
!pip install mlxtend --quiet

# 🧰 Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
from mlxtend.frequent_patterns import apriori, association_rules
from mlxtend.preprocessing import TransactionEncoder

# 📂 Load the dataset
from google.colab import files
uploaded = files.upload()  # Upload Groceries_dataset.csv

# 📑 Read the dataset
data = pd.read_csv('Groceries_dataset.csv')

# 🔄 Preprocessing
transactions = data.groupby('Member_number')['itemDescription'].apply(list).values.tolist()

# 🔁 Transforming dataset for Apriori
te = TransactionEncoder()
te_ary = te.fit(transactions).transform(transactions)
df = pd.DataFrame(te_ary, columns=te.columns_)

# ✅ Apply Apriori Algorithm
frequent_itemsets = apriori(df, min_support=0.02, use_colnames=True)
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1.0)

# 📊 Top Frequent Itemsets
print("Top 10 Frequent Itemsets:")
print(frequent_itemsets.sort_values(by='support', ascending=False).head(10))

# 📌 Association Rules Summary
print("\nAssociation Rules (Top 10):")
print(rules[['antecedents', 'consequents', 'support', 'confidence', 'lift']].head(10))

# 📉 Plotting Top 10 Frequent Items
frequent_itemsets['length'] = frequent_itemsets['itemsets'].apply(lambda x: len(x))
top_items = frequent_itemsets[(frequent_itemsets['length'] == 1)]
top_items = top_items.sort_values(by='support', ascending=False).head(10)

plt.figure(figsize=(10,6))
plt.bar(x=[', '.join(item) for item in top_items['itemsets']], height=top_items['support'], color='skyblue')
plt.title('Top 10 Frequent Items')
plt.xticks(rotation=45)
plt.xlabel('Items')
plt.ylabel('Support')
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```
